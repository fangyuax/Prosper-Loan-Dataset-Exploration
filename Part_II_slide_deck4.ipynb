## Investigation Overview


> The Prosper Loan Dataset presents a wide range of viewpoints on loan borrowing situation. In this investgation, I would like to find out what factors contribute to the the level of loans. This dataset helps us to learn about borrowers' backgrounds and how their credit rating and prosper rate effect the amount of their loan and the level of interest rate by processing data, visualizing data, and evaluating various graphs. In general, credit rating and prosper rate have a significant impact on the number of borrowers and interest rate.


## Dataset Overview

>  This dataset contains 113,937 loans with 81 variables on each loan, including loan amount, borrower rate (or interest rate), current loan status, borrower income, and many others. Most of the data is numeric, and there is some data type is datetime. Some variables such as CreditGrade, LoanStatus are ordered factor variables with the following levels.

## Visualization 1: Correlation between Loan Original Amount, Borrower Rate and Prosper Rating

> This heatmap displays that borrower rate and prosper rate have a strong correlation,which is -0.95, almost close to 1. This mean, the higher prosper rate that borrowers have, the lower borrower rate they have. Moreover, prosper rate has stronger correlation with loan original amount than borrower rate, which indicates that prosper rate regard as a more important factor that determine the loan original amount of the borrowers.

## Visualization 2: Relationship between Borrower Rate and Loan Original Amount with different credit grade borrowers

> In this FacetGrid, we could see that borrowers with higher credit grades such as AA or A generally have a lower interest rate for their loan. Most of the loan interest rates of AA and A grades are concentrated between 0.04-0.2, which are very low interest rate. When the credit grade is B, the interest rate begins to distribute above 0.2, and the distribution of 0.3 interest rate is more than the former two credit grade (A and AA). At the C, D, and E levels, interest rate range increses and the distribution becomes less concentrated. Some borrowers' interest rate reach higher than 0.35.

## Visualization 3: Relationship between Borrower Rate and Loan Original Amount with different prosper rating borrowers
>The trend becomes more apparent when we compare the relationship between the prosper rating and the interest rate for different borrowers. Borrowers with higher prosper rates have very low interest rates no matter how much they borrow. For example, the interest rate of borrowers with a prosper rate of 7.0 is almost kept below 0.12. The lower the borrower's prosper rate, the higher their borrower rate. For example, for a borrower with a prosper rate of 1.0, the interest rate could reach higher than 0.35, and there is no interest rate provided lower than 0.18. It can also be found that a borrower with a low prosper rate can borrow less loan than a borrower with a high prosper rate.
